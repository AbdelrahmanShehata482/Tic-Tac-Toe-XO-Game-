```python
# To print the board
def printBoard(board):
    print(board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('---------')
    print(board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('---------')
    print(board[7] + ' | ' + board[8] + ' | ' + board[9])
    print("\n")

player = 'O'
computer = 'X'
```

```python
# To check the position is free or not
def spaceIsFree(position):
    if board[position] == ' ':
        return True
    else:
        return False
```

```python
# To check if computer Win():
def checkcomputerWin():
    # True to end the game
    # False to continue

    # Condition for computer to win
    # for horizontal(win)
    if board[1] == 'X' and board[2] == 'X' and board[3] == 'X':
        return True
    elif board[4] == 'X' and board[5] == 'X' and board[6] == 'X':
        return True
    elif board[7] == 'X' and board[8] == 'X' and board[9] == 'X':
        return True

    # for diagonal(win)
    elif board[1] == 'X' and board[5] == 'X' and board[9] == 'X':
        return True
    elif board[3] == 'X' and board[5] == 'X' and board[7] == 'X':
        return True

    # for vertical(win)
    elif board[1] == 'X' and board[4] == 'X' and board[7] == 'X':
        return True
    elif board[2] == 'X' and board[5] == 'X' and board[8] == 'X':
        return True
    elif board[3] == 'X' and board[6] == 'X' and board[9] == 'X':
        return True
    return False
```

```python
# To check if player Win():
def checkplayerWin():
    # for horizontal(win)
    if board[1] == 'O' and board[2] == 'O' and board[3] == 'O':
        return True
    elif board[4] == 'O' and board[5] == 'O' and board[6] == 'O':
        return True
    elif board[7] == 'O' and board[8] == 'O' and board[9] == 'O':
        return True

    # diagonal(win)
    elif board[1] == 'O' and board[5] == 'O' and board[9] == 'O':
        return True
    elif board[3] == 'O' and board[5] == 'O' and board[7] == 'O':
        return True

    # vertical(win)
    elif board[1] == 'O' and board[4] == 'O' and board[7] == 'O':
        return True
    elif board[2] == 'O' and board[5] == 'O' and board[8] == 'O':
        return True
    elif board[3] == 'O' and board[6] == 'O' and board[9] == 'O':
        return True
    return False
```

```python
# To check if com or player draw
def checkDraw():
    for key in board.keys():
        if board[key] == ' ':
            return False
    return True
```

```python
def checkWhichMarkWon(mark):
    if board[1] == board[2] and board[1] == board[3] and board[1] == mark:
        return True
    elif (board[4] == board[5] and board[4] == board[6] and board[4] == mark):
        return True
    elif (board[7] == board[8] and board[7] == board[9] and board[7] == mark):
        return True
    elif (board[1] == board[4] and board[1] == board[7] and board[1] == mark):
        return True
    elif (board[2] == board[5] and board[2] == board[8] and board[2] == mark):
        return True
    elif (board[3] == board[6] and board[3] == board[9] and board[3] == mark):
        return True
    elif (board[1] == board[5] and board[1] == board[9] and board[1] == mark):
        return True
    elif (board[7] == board[5] and board[7] == board[3] and board[7] == mark):
        return True
    else:
        return False
```

```python
def insertLetter(letter, position):
    if spaceIsFree(position):
        board[position] = letter
        printBoard(board)
        if (checkDraw()):
            exit()
        elif checkWhichMarkWon(letter):    
            exit()

    else:
        print("Can't insert there!")
        position = int(input("Please enter new position:  "))
        insertLetter(letter, position)
```

```python
def player1Move():
    position = int(input("Enter the position for 'O':  "))
    insertLetter('O', position)
```

```python
def player2Move():
    position = int(input("Enter the position for 'X':  "))
    insertLetter("X", position)
```

```python
def compMove():
    bestScore = -700
    bestMove = 0
    for key in board.keys():
        if (board[key] == ' '):
            board[key] = computer
            score = minimax(board, 0, False)
            board[key] = ' '
            if (score > bestScore):
                bestScore = score
                bestMove = key
    insertLetter(computer, bestMove)
```

```python
def minimax(board, depth, isMaximizing):
    if (checkWhichMarkWon(computer)):
        return 1
    elif (checkWhichMarkWon(player)):
        return -1
    elif (checkDraw()):
        return 0

    if (isMaximizing):
        bestScore = -700
        for key in board.keys():
            if (board[key] == ' '):
                board[key] = computer
                score = minimax(board, depth + 1, False)
                board[key] = ' '
                if (score > bestScore):
                    bestScore = score
        return bestScore

    else:
        bestScore = 700
        for key in board.keys():
            if (board[key] == ' '):
                board[key] = player
                score = minimax(board, depth + 1, True)
                board[key] = ' '
                if (score < bestScore):
                    bestScore = score
        return bestScore
```
